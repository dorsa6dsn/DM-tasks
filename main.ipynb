پنج الگوریتم مناسب برای Binary Classification را معرفی کنید.

---

Logistic Regression:
احتمال تعلق یک نمونه به کلاس یک را حساب میکند مثلا احتمال برابر با 0.95باشد یعنی بیمار است ولی 0.32یعنی بیمار نیست
KNN:
معروف به تنبل ترین الگوریتم که با نگاه کردن به همسایه های خود تصمیم میگیرد و هرکلاسی که بیشتر تکرار شده باشد را انتخاب میکند
Desision Tree:
مثل بیست سوالی عمل میکند و سوال های متوالی میپرسد تا به نتیجه برسد
Random Forest:
مجموعه ای از درخت های تصمیم گیری است و به این گونه کار می کند که نتیجه ی 100 درخت را چک میکند
اگر 80درخت نتیجه داده باشند که بیمار است پس رای با اکثریت است
SVM:
سعی میکند بهترین مرز تصمیم گیری را بین دو کلاس در نظر بگیرد
خطی را در نظر میگیرد که دو گروه را با بیشترین فاصله از هم جدا کند
Naive Bayes:
بر اساس احتمال کار میکند و برای داده های متنی به شدت خوب کار میکند
مثلا قصد داریم تشخیص دهیم ایا این ایمیل اسپم است را خیر

که مطمئن شویم مدل الگو ها را یادگرفته نه که حفظ کرده باشد
در trainاز داده ها استفاده میکنیم برای اموزش مدل
در testمدل را ارزیابی میکنیم که اگر داده جدید به ان بدهیم درست کار میکند یا نه

## سوال 42

خطر استفاده از کل داده برای آموزش چیست؟

---

ممکن است مدل صرفا جواب ها را حفظ کرده باشد و وقتی داده جدید به ان بدهیم عملکرد خوبی نداشته باشد

## سوال 43

مفهوم Generalization چیست؟

---

معنی لغوی ان تعمیم است و به معنی توانایی مدل برای ارائه پیش بینی درست است
یعنی اگر مدل بعد از اموزش بتواند عملکرد خوبی داشته باشد یعنی تعمیم خوبی دارد

## سوال 44

مفهوم Overfitting چیست؟

---

زمانی که مدل بیش از اندازه به داده های اموزشی وابسته شود
در این حالت احتمال دارد حتی نویزهارا حساب کند و یادبگیرد

## سوال 45

مفهوم Underfitting چیست؟

---

نقاطه مقابل overfitting است به این معنی که داده های اموزش کم بوده و مدل خوب اموزش ندیده و عملکرد ان پایین است
