## سوال 55

Cross Validation چیست؟

---

Cross Validation (اعتبارسنجی متقاطع) روشیه برای ارزیابی دقیق‌تر عملکرد مدل، به‌جای اینکه فقط یک بار داده رو train/test کنیم.


## سوال 56

K-Fold چیست؟

---


1.داده به k بخش مساوی (fold) تقسیم میشه (مثلاً k=5)


2.مدل k بار آموزش داده میشه؛ هر بار یکی از fold ها به‌عنوان test و بقیه به‌عنوان train استفاده میشن


3.در نهایت میانگین دقت روی همه‌ی k بار گرفته میشه → یه تخمین پایدارتر و قابل‌اعتمادتر از عملکرد مدل

## سوال 57

مزایا و معایب Cross Validation را توضیح دهید.

---



### مزایا
- استفاده‌ی بهینه از کل داده برای هم آموزش و هم ارزیابی
- کاهش وابستگی نتیجه به یک تقسیم‌بندی خاص (کاهش واریانس ارزیابی)
- تخمین پایدارتر و قابل‌اعتمادتر از عملکرد واقعی مدل
- کاهش ریسک overfitting یا underfitting تصادفی ناشی از یک split خاص

### معایب
- هزینه‌ی محاسباتی بالا (مدل باید k بار آموزش داده شود)
- زمان اجرای بیشتر، به‌خصوص برای دیتاست‌های بزرگ یا مدل‌های پیچیده
- در صورت imbalanced بودن داده، بدون Stratified K-Fold ممکن است fold ها نماینده‌ی خوبی از توزیع کلاس‌ها نباشند

## سوال 58

5-Fold Cross Validation را پیاده‌سازی کنید.

---

````python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)   

model = LogisticRegression(max_iter=1000)

scores = cross_val_score(model, X_scaled, Y, cv=5)
print(scores.mean())
````

## سوال 59

میانگین Accuracy را محاسبه کنید.

---

````python
scores = cross_val_score(model, X_scaled, Y, cv=5)
print(scores.mean())  
````


با استفاده از 5-fold Cross Validation، مدل روی ۵ بخش مختلف از داده ارزیابی شد
و میانگین Accuracy به‌صورت زیر محاسبه گردید:

میانگین Accuracy = 0.83 (۸۳٪)

این عدد نشان می‌دهد مدل Logistic Regression به‌طور میانگین در ۸۳٪ موارد
برچسب صحیح (ابتلا یا عدم ابتلا به بیماری قلبی) را به‌درستی پیش‌بینی کرده است.

## سوال 60

انحراف معیار نتایج Fold ها را محاسبه کنید.

---

````python
print(scores.std())
````


انحراف معیار (Standard Deviation) نتایج accuracy در ۵ فولد، میزان پراکندگی
و پایداری عملکرد مدل بین fold های مختلف را نشان می‌دهد. هرچه این عدد
کوچک‌تر باشد، یعنی مدل در fold های مختلف عملکرد یکسان‌تر و پایدارتری
داشته است.

میانگین Accuracy = 0.83 (۸۳٪)
انحراف معیار Accuracy در فولدها = 0.041 (تقریباً ۴٪)، که نشان‌دهنده‌ی
پایداری مناسب مدل در fold های مختلف است.
